```python
from pydantic import BaseModel, Field, ValidationError
from typing import Literal
```

```python
# Input model
class BibleReference(BaseModel):
    book: str = Field(min_length=1, max_length=50, description="Bible book name")
    chapter: int = Field(ge=1, description="Chapter number (≥1)")
    verse: int = Field(ge=1, description="Verse number (≥1)")
    language: Literal["english", "tamil"] = "english"
    version: str = "kjv"
```

```python
# Output model
class VerseResponse(BaseModel):
    reference: str  # "John 3:16"
    version: str    # "KJV"
    language: str   # "english" or "tamil"
    text: str       # The actual verse text
```

```python
# User input
try:
    book = input("Enter book name: ").strip()
    chapter = int(input("Enter chapter: "))
    verse = int(input("Enter verse: "))

    lang_input = input("Language (english/tamil): ").strip().lower()
    language = lang_input if lang_input in ["english", "tamil"] else "english"

    # Validate input
    ref = BibleReference(
        book=book,
        chapter=chapter,
        verse=verse,
        language=language
    )

    print(f"✓ Valid: {ref.book} {ref.chapter}:{ref.verse} ({ref.language})")

except ValidationError as e:
    print(f"✗ Invalid:\n{e}")
```
